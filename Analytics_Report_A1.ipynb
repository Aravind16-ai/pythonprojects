```python
# Data Cleaning
# Handling missing values
prod_df.dropna(inplace=True)

# Handling outliers (optional, depending on your data)
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
# Time series analysis of production rates over time
import matplotlib.pyplot as plt

# Plot production rates over time
plt.figure(figsize=(12, 6))
plt.plot(meta_df['FirstProdDate'], meta_df['PeakProd_BOE'], label='Peak Production (BOE)')
plt.plot(meta_df['FirstProdDate'], meta_df['CumProd_BOE'], label='Cumulative Production (BOE)')
plt.xlabel('Date')
plt.ylabel('Production (BOE)')
plt.title('Production Over Time')
plt.legend()
plt.show()
```

## Key obseravation: 
The time series analysis depicts the evolution of peak and cumulative production rates over time. The graph visualizes production trends, providing insights into production dynamics and long-term performance. Stakeholders can leverage this analysis to evaluate production efficiency and make informed decisions regarding resource management and investment strategies.

```python
# Set the threshold for correlation
corr_threshold = 0.5  # Adjust as needed

# Calculate the correlation matrix
correlation_matrix = meta_df[['PeakProd_BOE', 'CumProd_BOE']].corr()

# Select only the columns with a high correlation
corr_features = correlation_matrix[((correlation_matrix >= corr_threshold) | (correlation_matrix <= -corr_threshold)) & (correlation_matrix != 1.0)]
variables = corr_features.loc[corr_features.sum() != 0.0, corr_features.sum() != 0.0].columns

# Check if variables list is not empty
if len(variables) > 0:
    # Plot the regression plots for those features
    plot = sns.pairplot(data=meta_df[variables], kind='reg', plot_kws=dict(scatter_kws={'color': 'g', 'alpha': 0.3, 's': 3}, line_kws={'color': 'red'}), height=5)

    # Annotate with the correlation values
    for i in range(len(plot.axes)):
        for j in range(len(plot.axes[i])):
            # For the diagonal, there are not always labels - skip those
            if i != j:
                y_label = plot.axes[i][j].get_ylabel()
                x_label = plot.axes[i][j].get_xlabel()
                plot.axes[i][j].text(0.5, 0.5, f"corr: {correlation_matrix.loc[x_label, y_label]:.2f}", size=18, ha='center', va='center', transform=plot.axes[i][j].transAxes)

    # Show the plots
    plt.show()
else:
    print("No variables found for pair plot.")
```

## Key observation:
The correlation analysis examines the relationship between peak and cumulative production volumes. With a correlation threshold set at 0.5, the correlation matrix highlights highly correlated variables. Pair plots are generated for variables meeting the threshold, illustrating their linear relationship. The correlation coefficients annotated on each plot provide insights into the strength and direction of correlations. This analysis aids in identifying potential dependencies between production metrics, guiding strategic decision-making processes for optimizing production operations.

```python
# Select relevant cost metrics columns from meta_df
cost_metrics_columns = ['TotalWellCost_USDMM', 'DrillingCost_USDMM']

# Calculate correlation matrix
cost_metrics_corr = meta_df[cost_metrics_columns].corr()

# Visualize correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cost_metrics_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Cost Metrics Correlation Matrix')
plt.show()
```

## Key observation:
A correlation matrix is computed for selected cost metrics columns from the dataset. The matrix quantifies the linear relationship between total well cost and drilling cost, providing insights into their correlation strength and direction. The heatmap visualization enhances interpretation by color-coding correlation values, enabling quick identification of significant correlations. This analysis aids decision-making in cost optimization and resource allocation within the oil and gas industry.

```python
# Calculate correlations
correlations = cost_metrics_corr

# Define correlation threshold
corr_threshold = 0.5

# Select only the columns with a high correlation
corr_features = cost_metrics_corr[((correlations >= corr_threshold) | (correlations <= -corr_threshold)) & (correlations != 1.0)]
variables = corr_features.loc[corr_features.sum() != 0.0, corr_features.sum() != 0.0].columns

# Plot the regression plots for those
plot = sns.pairplot(data=meta_df[variables],
                    kind='reg',
                    plot_kws=dict(scatter_kws={'color': 'g', 'alpha': 0.3, 's': 3}, line_kws={'color': 'red'}),
                    height=5)

# Annotate with the correlation values
for i in range(len(plot.axes)):
    for j in range(len(plot.axes[i])):
        # For the diagonal there are not always labels - skip those
        if i != j:
            y_label = plot.axes[i][j].get_ylabel()
            x_label = plot.axes[i][j].get_xlabel()
            plot.axes[i][j].text(0.5,
                                 0.5,
                                 f"corr: {correlations.loc[x_label, y_label]:.2f}",
                                 size=18,
                                 ha='center',
                                 va='center',
                                 transform=plot.axes[i][j].transAxes)

# Show the plot
plt.show()
```

## Key observation:
A threshold-based approach is employed to identify highly correlated cost metrics. Correlations exceeding a specified threshold (e.g., 0.5) are visualized through regression plots, illustrating the relationship between variables. Scatter plots with regression lines highlight correlation strength, aiding in understanding cost interdependencies. Annotation of correlation coefficients enhances interpretation, facilitating decision-making in resource allocation and cost management within the oil and gas sector.
